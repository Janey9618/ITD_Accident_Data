<H2>Use of One way Annova</H2>

```python
pip install scipy
```

```python
import scipy
```

```python
from scipy.stats import f_oneway
```

<h3>Testing correlation between body mass and sex:</h3>

```python
( penguins[penguins['sex'] == 'male']['body_mass_g'],
 penguins[penguins['sex'] == 'female']['body_mass_g'])
```

```python
f_stats, p_value = f_oneway( penguins[penguins['sex'] == 'male']['body_mass_g'],
                             penguins[penguins['sex'] == 'female']['body_mass_g'])

p_value
```

<h2>Insight #15</h2>
<h3>There is no correlation between the sex and the body mass</h3>

<h3>Testing correlation between body mass and island:</h3>

```python
( penguins[penguins['island'] == 'dream']['body_mass_g'],
 penguins[penguins['island'] == 'torgersen']['body_mass_g'],
 penguins[penguins['island'] == 'biscoe']['body_mass_g'])
```

```python
f_stats, p_value = f_oneway( penguins[penguins['island'] == 'dream']['body_mass_g'],
                             penguins[penguins['island'] == 'torgersen']['body_mass_g'],
                             penguins[penguins['island'] == 'biscoe']['body_mass_g'])


p_value
```

<h2>Insight #16</h2>
<h3>There is no correlation between island and body mass</h3>
